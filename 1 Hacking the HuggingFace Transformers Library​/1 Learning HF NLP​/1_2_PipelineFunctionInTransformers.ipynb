Example of using the pipeline function for sentiment analysis:

```python
from transformers import pipeline

# Load sentiment analysis pipeline
sentiment_analysis = pipeline("sentiment-analysis")

# Provide input text
text = "I love using this product. It's amazing!"

# Get sentiment analysis result
result = sentiment_analysis(text)

# Print the result
print(result)
```

Example using the text-generation pipeline:

```python
from transformers import pipeline

# Load text generation pipeline
text_generator = pipeline("text-generation")

# Provide a prompt for text generation
prompt = "Once upon a time in a"

# Generate text based on the prompt
generated_text = text_generator(prompt, max_length=50, num_return_sequences=1)

# Print the generated text
print(generated_text[0]['generated_text'])
```
